```python
import re
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from tqdm import tqdm  

# Firefox 드라이버 로드
driver = webdriver.Firefox()

# 결과를 저장할 리스트
supporters = []
reward_levels = []
likes = []
shares = []
prices = []
achievement_rate = []
achievement_goals = []
funding_periods = []
urls = []

for _, row in tqdm(df_2.iterrows(), total=df_2.shape[0]):
    url = row['url']
    


    # URL에 접속
    driver.get(url)
    time.sleep(2)
         
    #left join 
    urls.append(url)

    # 페이지 HTML 파싱
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # 후원자 수 크롤링
    supporters_label = soup.find('div', class_='ProjectIntroduction__StatusTitle-sc-7frh7h-13 jDLdcH', text='후원자')
    if supporters_label is not None:
        num_supporters = supporters_label.find_next_sibling().text
        supporters.append(num_supporters)
    else:
        supporters.append("0")

    # 좋아요 수 크롤링
    likers_label = soup.find("div", class_="DetailLikeButton__StyledDetailLikeButton-v1ofbi-1 dOAQhO")
    span_element_likes = likers_label.find('span') if likers_label else None
    if span_element_likes is not None:
        num_likers = span_element_likes.get_text(strip=True)
        likes.append(num_likers)
    else:
        likes.append("0")
    
    # 공유 수 크롤링
    shares_label = soup.find('button', class_='PledgeButton__StyledShareSNSButton-sne71z-4')
    span_element_shares = shares_label.find('span') if shares_label else None
    if span_element_shares is not None:
        num_shares = span_element_shares.get_text(strip=True)
        shares.append(num_shares)
    else:
        shares.append("0")
    
    # 펀딩 목표 금액, 펀딩 기간 크롤링
    achievement_goal = soup.find_all('dd', re.compile('FundingInformation__Desc'))[0].text
    achievement_goal = int(re.sub('\D','',achievement_goal))
    achievement_goals.append(achievement_goal)
    funding_period = soup.find_all('dd', re.compile('FundingInformation__Desc'))[1].text
    funding_period = re.sub('[가-힣]', '', funding_period)
    funding_periods.append(funding_period)

    # 달성금액, 달성률 크롤링
    price_achievement = soup.find_all('div', re.compile('ProjectIntroduction__StatusValue'))[0].text
    price = price_achievement.split('원')[0]
    prices.append(price)
    achievement = price_achievement.split('원')[1]
    achievement_rate.append(achievement)

    # 선물 금액 크롤링
    driver.execute_script("window.scrollBy(0, 1000);")
    time.sleep(1.5)  # wait for the page to load
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    reward_levels_elements = soup.find_all(class_="RewardSelector__RewardMinimumPledgeAmount-sc-10bnmkm-1 iUCmew")
    page_values = [element.text for element in reward_levels_elements]
    reward_levels.append(page_values)

# 데이터프레임으로 저장
df_result = pd.DataFrame({
    'urls':urls,
    'supporters': supporters,
    'reward_levels': reward_levels,
    'like_num': likes,
    'share_num': shares,
    'price': prices,
    'achievement_rate': achievement_rate,
    'achievement_goal': achievement_goals,
    'funding_period': funding_periods
})

# 드라이버 종료
driver.quit()
```
