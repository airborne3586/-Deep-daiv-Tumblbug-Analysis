```python
import pandas as pd 

df_app = pd.read_csv('df_app_results.csv',index_col = 0)
df_hobby = pd.read_csv('df_hobby_results.csv',index_col = 0)
df_magazine = pd.read_csv('df_magazine_results.csv',index_col = 0)
```

```python
print(df_magazine.head())
print(df_hobby.head())
print(df_app.head())
```

# 전처리 

```python
df_hobby['Reward Levels'] = df_hobby['Reward Levels'].str.replace(',','')
df_hobby['Reward Levels'] = df_hobby['Reward Levels'].str.replace('원', '')
df_hobby['Reward Levels'] = df_hobby['Reward Levels'].str.replace('+', '').astype(float)

df_app['Reward Levels'] = df_app['Reward Levels'].str.replace(',','')
df_app['Reward Levels'] = df_app['Reward Levels'].str.replace('원', '')
df_app['Reward Levels'] = df_app['Reward Levels'].str.replace('+', '').astype(float)

print(df_hobby.head())
print(df_app.head())
```

# 시각화

3가지 카테고리에 대해 boxplot을 시각화 해보고 이상치를 어떻게 처리할 것인지 확인한다

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18, 6))

# df_app 박스플롯
plt.subplot(1, 3, 1)
sns.boxplot(x=df_app['Reward Levels'], color='skyblue')
plt.title('df_app Reward Levels 박스플롯')

# df_magazine 박스플롯
plt.subplot(1, 3, 2)
sns.boxplot(x=df_magazine['Reward Levels'], color='skyblue', orient='h')
plt.title('df_magazine Reward Levels 박스플롯')

# df_hobby 박스플롯
plt.subplot(1, 3, 3)
sns.boxplot(x=df_hobby['Reward Levels'], color='skyblue', orient='h')
plt.title('df_hobby Reward Levels 박스플롯')

plt.tight_layout()
plt.show()
```
